We use Anova test,If we have more than two samples.Analysis of Variance.

**One-Way Anova Test:**
This is used to compare the mean various group based on one independent variable that is factor.

H0 is All Groups have same Mean.

Ha is Atleast one group have different mean.

```python
df.dose.unique()
```

```python
from scipy import stats #To perform One-way Anova
f,p=stats.f_oneway(df[df['dose']==0.5]['len'],df[df['dose']==1]['len'],df[df['dose']==2]['len'])
print(f,p)
```

```python
p_round=round(p,1)
p_round
```

```python
alpha=0.05
if p<0.05:
    print(p_round,"Reject Null Hypothesis,Atleast one group have different mean")
else:
    print(p_round,"Accept null hypothesis,All Groups have same Mean")
```

**Two-Way Anova Test:**
This is used to examine two independent variables or factors based on one dependent variance and to find the interactions between these factors.

H0 is There is no difference in tooth growth between different doses of Vitamin C.
There is no difference in tooth growth between different types of supplements.
There is no interaction between dose and supplement type on tooth growth.

Ha is Atleast one of the null hypothesis is false.

```python
import statsmodels.api as sm # To permorm Two way anova
from statsmodels.formula.api import ols 
#Model we're going to use for two way anova-ordinary least square model
```

Based on the flexibility and capabilities of the OLS framework,we're using it for two-way anova test.some of them are:

Parameter Estimation,Categorical Encoding,F-statistic Calculation,P-values calculation,Modeling Interactions etc..,

```python
#Fit the ols Model
model=ols('len~C(supp)+C(dose)+C(supp):C(dose)',data=df).fit()
```

C represent categorical variable,
~ represents independent variable,
: represent the interaction between two independent variables

len ~ C(supp) + C(dose) + C(supp):C(dose) allows us to analyze the effects of both supp and dose, as well as their interaction, on the dependent variable len.

```python
#Two way Anova
anova=sm.stats.anova_lm(model,typ=2)
```

```python
print(anova)
```

If you observe the p-values of C(supp),C(dose),C(supp):C(dose),all are less than significance level 0.05.so Reject the null Hypothesis.

Based on your research you have to choose the one-way or two-way anova test.

If you are investigating the effect of a single independent variable (factor) on a dependent variable or you want to compare the means of three or more independent groups then one-way is the perfect choice!

If you are investigating the effect of two independent variables (factors) on a dependent variable or you want to understand if there is an interaction effect between the two factors then two-way is the perfect choice!
