Let's Learn about t test,z test and Proportion test in this tutorial.

**T test**

Basically in hypothesis testing we have to select test statics based on the distribution and type of the data.In the case of t test,If the given data is continuous and we do not know the pupulation standard deviation then we need to go with t test.
Let me consider an example to understand the basic concept.

**One Tail Test:**
If Alternate hypothesis contains < or >.

Example:An automobile tyre manufacturer claims that the average life of a particular grade of tyre is more than 18,000 km. A random sample of 15 tyres is having mean 20,000 km with a standard deviation of 4000 km.

Here H0 is Mu<=18,000 #No action/difference

     H1 is Mu>18,000

x_=20000
n=15
std=4000

```python
t=(20000-18000)/(4000/np.sqrt(15))
print(t)
```

```python
from scipy import stats
p=stats.t.cdf(t,14)
alpha=0.05 #By default the value of significance level is 0.05
if p>alpha:
    print(p,"The average life of a particular grade of tyre is not more than 18,000 km")
else:
    print(p,"The average life of a particular grade of tyre is more than 18,000 km")
```

**Two Tail Test:**
If Alternate Hypothesis contains = or !=

Example:An automobile tyre manufacturer claims that the average life of a particular grade of tyre is equal to 18,000 km. A random sample of 15 tyres is having mean 20,000 km with a standard deviation of 4000 km.

Here H0 is Mu!=18,000

     H1 is Mu=18,000

x_=20000
n=15
std=4000

```python
p=2*stats.t.cdf(t,df=14)
p
```

```python
alpha=0.05 #By default the value of significance level is 0.05
if p>alpha:
    print(p,"The average life of a particular grade of tyre is not equal to 18,000 km")
else:
    print(p,"The average life of a particular grade of tyre is equal to 18,000 km")
```

Z Test:

In the case of z test,If the given data is continuous and we know the pupulation standard deviation then we need to go with z test.z-test assumes that the test statistic (z-score) follows a standard normal distribution.

Let me consider an example to understand the basic concept.

Example: A school claimed that the students who are studying in his school that are more intelligent than the average school. On calculating the IQ scores of 50 students, the average turns out to be 110. The mean of the population IQ is 100 and the standard deviation is 15. State whether the claim of the principal is right or not at a 5% significance level.

Here,H0 is Mu<=100
           Mu>100
           
x_=100
Mu=100
Sigma=15
n=50

```python
z=(110-100)/(15/np.sqrt(50))
round(z,2)
```

```python
p=1-stats.norm.cdf(z)
p
```

```python
alpha=0.05
if p>alpha:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

**Proportion Test:**
The share of a population that belongs to a particular category.

Example:A study is stated that 40% Nobel Prize winners were born in the UK.

We will take a sample of 50 randomly selected Nobel Prize winners ,we could find that 15 out of 50 Nobel Prize winners in the sample were born in the UK.

Here,H0 is 40% Nobel Prize winners were not born in the UK
     H1 is 40% Nobel Prize winners were born in the UK

The sample proportion p^ = 15/50 = 0.30 = 30 %
The population proportion p= 0.40 = 40%

```python
ts=(0.3-0.4)/(np.sqrt((0.4*(1-0.4))/(50)))
ts
```

```python
print(stats.norm.cdf(ts))
```

```python
alpha=0.05
if p>alpha:
    print("Accept Null Hypothesis")
else:
    print("Reject Null Hypothesis")
```

**Two Sample Test:**
If we have two different samples then we need to do two sample test.

Let me explain by taking two different series.

```python
A=pd.Series([91, 87, 99, 77, 88, 91])
B=pd.Series([101, 110, 103, 93, 99, 104])
```

```python
stats.ttest_ind(A,B) #Two samples are independent in 2 sample test
```

```python
p=stats.ttest_ind(A,B)[1]
p
```

These are the basic Hypothesis testing methods.In any test we can compare the obtailed p value with significance level or we can compare critical value with test statitics to to reject or accept Null Hypothesis.
