We perform Chi-square test on categorical variables.
Let's perform chi2 test on two categorical variables in titanic data set.Survived and Pclass.

H0 is No significant association between Survived and Pclass.

Ha is there is a significant association between Survived and Pclass.

```python
from scipy.stats import chi2_contingency #To perfom Chi2-test
```

```python
contigency_table=pd.crosstab(df['Survived'],df['Pclass'])
```

```python
chi2,p,df,expected=chi2_contingency(contigency_table)
```

```python
print(p)
```

```python
if p<0.05:
    print(p,"Reject Null Hypothesis,There is a significant association between Survived and Pclass.")
else:
    print(p,"Accept Null Hypothesis,There is no significant association between Survived and Pclass.")
```
